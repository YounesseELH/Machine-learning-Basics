**Exercice 1: function qui permet de creer une tableau de (m,n+1) dimension on ajoutons une colonne des 1**

```python
def initialisation(m,n):
  M = np.random.randn(m,n)
  #print(M)
  M1 = np.ones((1,m)).T
  #print(M1)
  return np.hstack((M,M1))
```

```python
A = initialisation(3,3)
A
```

**Exerice 2: faire une zoom sur une image**

```python
from scipy import misc 
import matplotlib.pyplot as plt
face = misc.face(gray=True)
plt.imshow(face,cmap=plt.cm.gray)
```

```python
face.shape
```

```python
h = face.shape[0]
w = face.shape[1]
print(h,w)
```

```python
image_avec_zoom = face[h//4:-h//4,w//4:-h//4]
plt.imshow(image_avec_zoom,cmap=plt.cm.gray)
```

```python
#image avec mois qualite
image_avec_zoom = face[::10,::10]
plt.imshow(image_avec_zoom,cmap=plt.cm.gray)
```

```python
def f(x):
  return x**2+1

def g(x):
  return x**3-1
#generer 200 number entre -3 et 3:
x = np.linspace(-3,3,200)
plt.plot(x,f(x), x,g(x))
plt.show()
```

**Scatter :  pour tracer une nuage de points**

```python
x = np.linspace(0,10,30)
y = np.linspace(0,10,30)
plt.scatter(x,y,c='red') #c: pour le couleur
plt.show()
```

**Cycle de vie**

```python
plt.figure(figsize=(12,8))
plt.plot(x,y)
plt.plot(x**2,y)
plt.title('Figure 1')
plt.xlabel('Axe x')
plt.ylabel('Axe y')
plt.show()
```

**Exercice:**

```python
dataset = {f"experience{i}":np.random.randn(100) for i in range(4)}
```

```python
#dataset
```

```python
def graphique(data):
  n = len(data)
  plt.figure(figsize=(12,8))
  for i,j in zip(data.keys(), range(1,n+1)):
    plt.subplot(n,1,j)
    plt.plot(data[i])
    plt.title(i)

  plt.show()
```

```python
graphique(dataset)
```
